```python
# Count the occurrences of each crime type
# Exclude 'Unit Name' and 'Year' as they are not crime types
crime_columns = df.columns.drop(['Unit Name', 'Year', 'Total Cases'])
crime_type_counts = df[crime_columns].sum().sort_values(ascending=False)
print(crime_counts)
```

What are the most common types of crime?

```python
#Create the bar plot
plt.figure(figsize=(12, 7))
sns.barplot(x=crime_type_counts.values, y=crime_type_counts.index)
plt.title('Top 10 Most Common Crime Types', fontsize=16)
plt.xlabel('Number of Reported Cases')
plt.ylabel('Crime Type')

plt.show()
```

But what does this trend consist of? As the bar chart shows, the most prevalent crime is by far other types or mixed crimes, followed by Narcotics. This tells us what kind of crimes are most common on the ground.

Which areas have the highest reported crime rates?

```python
#claculate total crimes accoridng to Unit Name column
total_crimes_per_region = df.groupby('Unit Name')['Total Cases'].sum().sort_values(ascending=False)
print(total_crimes_per_region)
```

```python
plt.figure(figsize=(17, 8))
sns.barplot(x=total_crimes_per_region.head(10).index, y=total_crimes_per_region.head(10).values)
plt.title('Top 10 Regions with the Highest Reported Crimes', fontsize=16)
plt.xlabel('Region')
plt.ylabel('Number of Reported Crimes')
plt.show()
```

Resolution: Our data shows that while overall crime has increased, the most significant challenge is Other Cases and Narcotics, particularly in the Dhaka Range. This suggests we should enable the drug enforcement agency and drive against drug abuse in Dhaka more as well as with an eye for all other types of crime in and around Dhaka area.
